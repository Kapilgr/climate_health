# Handling outliers

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

```

```python
df.columns
```

```python
columns_to_skip = ['heat_wave_days','drought_indicator','extreme_weather_events']
for col in numeric_cols:
    if col not in columns_to_skip:
        df = remove_outliers(df, col)

   
```

# Creating new features based on domain knowledge to enrich the dataset. 

```python
# Adding the feature named "pollution_index"
df['pollution_index']=df['pm25_ugm3']*df['air_quality_index']
# Higher the pollution index, higher will be the risk of causing several diseases.

df['multi_disease_burden'] =df['respiratory_disease_rate'] + df['cardio_mortality_rate'] + df['vector_disease_risk_score'] + df['waterborne_disease_incidents']
# This indicates which region's people is vulnerable to multiple diseases

df['healthcare_gap'] = df['multi_disease_burden'] / df['healthcare_access_index']

# It quantifies the mismatch between health needs and healthcare availability.
# High disease burden + low healthcare access ->large healthcare gap
# Low disease burden + high healthcare access -> small healthcare gap
```

```python
# Plotting histogram
columns_to_plot=['temperature_celsius','temp_anomaly_celsius', 'precipitation_mm','respiratory_disease_rate','air_quality_index','waterborne_disease_incidents']
for col in df.columns:
    if col in columns_to_plot:
        plt.figure()
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()

       
```

- **Normal Distribution:** Temperature follows a typical bell curve pattern.
- **Moderate Climate Dominant:** Most frequent temperatures cluster in middle ranges.
- **Few Temperature Extremes:** Limited very hot or very cold occurrences.

Based on the temperature anomaly histogram, here are the key insights:
- **Normal Distribution:** Temperature anomalies follow a bell curve centered near 0°C.
- **Climate Stability:** Most periods experience minimal temperature deviation from normal.
- **Balanced Extremes:** Warming and cooling events occur with roughly equal frequency.
- **Decision-Making:** Current climate shows stability - maintain monitoring systems.

Based on the precipitation histogram, here are the key insights:
- **Right-Skewed Distribution:** Most periods have low-moderate rainfall (0-50mm).
- **Frequent Dry Spells:** High frequency of low precipitation events.
- **Rare Heavy Rainfall:** Extreme precipitation (>100mm) occurs infrequently.

Based on the air quality index histogram, here are the key insights:

- **Frequent Good Air:** High frequency of low AQI values indicating generally clean air.
- **Rare Poor Air Quality:** Few instances of unhealthy air levels (>150 AQI).
- **Decision-Making:** Focus pollution control on specific places rather than widespread measures.

Based on the respiratory disease rate histogram, here are the key insights:
- **Right-Skewed Distribution:** Most populations have low-moderate disease rates.
- **Generally Healthy:** Majority experience good respiratory health.
- **Vulnerable Subgroups:** Small segments show significantly higher disease burden.
- **Decision-Making:** Target healthcare interventions toward high-risk subgroups.

Based on the waterborne disease incidents histogram, here are the key insights:
- **Right-Skewed Distribution:** Most areas have few to no waterborne disease incidents.
- **Generally Safe Water:** Majority of locations show good water quality outcomes.
- **Localized Outbreaks:** Few areas experience concentrated disease clusters.
- **Decision-Making:** Focus sanitation efforts on identified high-incident places.



```python
# Scatter plot
bins = [0, 50, 100, 150, np.inf]
labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy']

df['air_quality_level'] = pd.cut(df['air_quality_index'], bins=bins, labels=labels)

sns.scatterplot(data=df, x='air_quality_index', y= 'pm25_ugm3',hue='air_quality_level')
plt.show()
```

# Key Insights
There is a strong positive linear relationship: as Air Quality Index increases, PM2.5 concentration also increases.
The color–coded categories show smooth transitions between "Good", "Moderate", "Unhealthy for Sensitive Groups", and "Unhealthy," confirming consistency between AQI bins and PM2.5 values.

PM2.5 rises almost proportionally with AQI.
At higher AQI levels (200+), PM2.5 values cluster tightly, indicating more consistent and severe pollution levels.

PM2.5 is a strong predictor of overall air quality — models predicting AQI can heavily rely on PM2.5.
As the color transitions are clean, it suggests AQI classification is consistent and reliable in this dataset.
Pollution control policies should target activities contributing to PM2.5 because reducing it will directly lower AQI.

Air Quality Index is a strong predictor of respiratory health outcomes, making it a key variable for public health planning.
